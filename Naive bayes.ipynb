```python
import pandas as pd
df=pd.read_csv("E:\\DS Data\\titanic.csv")
df.head()
```

```python
df.drop(['PassengerId',"SibSp","Name","Ticket","Cabin","Embarked"],axis='columns',inplace=True)
df.head(3)
```

```python
target=df.Survived
target.head(3)
```

```python
inputs=df.drop('Survived',axis='columns')
inputs.head(3)
```

```python
dummies=pd.get_dummies(inputs.Sex)
dummies.head()
```

```python
inputs=pd.concat([inputs,dummies],axis='columns')
inputs.head(3)
```

```python
inputs.drop(['Sex','male'],axis='columns',inplace=True)
inputs.head(3)
```

```python
inputs.columns[inputs.isna().any()]
```

```python
inputs.Age=inputs.Age.fillna(inputs.Age.mean())
```

```python
inputs.head(20)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(inputs,target,test_size=0.1)
```

```python
X_test
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test[:10])
```

```python
y_test[:10]
```

```python
model.score(X_test,y_test)
```

```python
model.predict_proba(X_test[:10])
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(GaussianNB(),X_train, y_train, cv=10)
```
