```python
df4=df3.copy()
```

```python
dummies=pd.get_dummies(df4.ExerciseAngina)
```

```python
df4=pd.concat([df4,dummies],axis='columns')
```

```python
df4.head()
```

```python
df4=df4.drop(['ExerciseAngina','N'],axis='columns')
df4.head()
```

```python
df4.ST_Slope.replace(
    {
        'Down': 1,
        'Flat': 2,
        'Up': 3
    },
    inplace=True
)

df4.RestingECG.replace(
    {
        'Normal': 1,
        'ST': 2,
        'LVH': 3
    },
    inplace=True)

df4.head()
```

```python
df5=pd.get_dummies(df4)
df5.head()
```

```python
X=df5.drop(("HeartDisease"),axis='columns')
y=df5.HeartDisease
```

```python
from sklearn.svm import SVC
model=SVC()
from sklearn.model_selection import cross_val_score
```

```python
scores = cross_val_score(model, X, y, cv=5)
scores.mean()
```

```python
## bagging
from sklearn.ensemble import BaggingClassifier
bagmodel=BaggingClassifier(base_estimator=model,
    n_estimators=10,
    max_samples=0.8,
    oob_score=True)
```

```python
from sklearn.model_selection import cross_val_score
scoresnew = cross_val_score(bagmodel, X, y, cv=5)
scoresnew.mean()
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
from sklearn.model_selection import cross_val_score
```

```python
score=cross_val_score(DecisionTreeClassifier(),X,y,cv=5)
score.mean()
```
